```python
def create_node(data, next_node=None):
    return {'data': data, 'next': next_node}

def create_queue():
    return {'front': None, 'rear': None}

def enqueue(queue, data):
    new_node = create_node(data)
    if not queue['front']:
        queue['front'] = new_node
        queue['rear'] = new_node
    else:
        queue['rear']['next'] = new_node
        queue['rear'] = new_node
    print(f"Enqueued {data}. Front: {queue['front']['data']}, Rear: {queue['rear']['data']}")
    return queue

def dequeue(queue):
    if not is_empty(queue):
        data = queue['front']['data']
        queue['front'] = queue['front']['next']
        if not queue['front']:
            queue['rear'] = None
        print(f"Dequeued {data}.")
        return data
    else:
        print("Cannot dequeue from an empty queue.")
        return None

def is_empty(queue):
    return queue['front'] is None

def peek(queue):
    if not is_empty(queue):
        return queue['rear']['data']
    else:
        print("Queue is empty.")
        return None

# Example usage
queue = create_queue()
for i in range(10):
    enqueue(queue, i)
for j in range(2):
    dequeue(queue)

print("Peek: rear: ", peek(queue))
print("Peek: queue: ",queue)
```
